```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# -------------------------
# 1. Define input qubits
# -------------------------
# 2 qubits + 2 classical bits
qc = QuantumCircuit(2, 2)

# Input state |00>
qc.barrier()
qc.h(0)      # Hadamard gate on qubit 0
qc.x(1)      # Pauli-X gate on qubit 1

qc.barrier()

# -------------------------
# 2. Apply single-qubit gates
# -------------------------
qc.y(0)      # Pauli-Y
qc.s(1)      # Phase gate
qc.t(0)      # T gate

qc.barrier()

# -------------------------
# 3. Apply two-qubit gate
# -------------------------
qc.cx(0, 1)  # CNOT (control: q0, target: q1)

qc.barrier()

# -------------------------
# 4. Measure output qubits
# -------------------------
qc.measure(0, 0)
qc.measure(1, 1)

# -------------------------
# 5. Draw the circuit
# -------------------------
print(qc.draw())

# -------------------------
# 6. Run on Aer simulator
# -------------------------
backend = Aer.get_backend('aer_simulator')
result = execute(qc, backend, shots=1024).result()
counts = result.get_counts()

# -------------------------
# 7. Show output
# -------------------------
print("Output counts:", counts)
plot_histogram(counts)
plt.show()
```
